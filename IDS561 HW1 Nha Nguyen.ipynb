```python
#Import libraries
import pandas as pd
import numpy as np

#Reading Text File 
with open('temperatures.txt') as f: 
    lines = f.readlines()

class BDHW:
    
    def converting_listofintegers(self, lines):
        y = [] 
        z = []
        
#Create a list of all given 1000 lines -string within list 
        x = [i.split('\n')[0].strip("()").split("), (") for i in lines]            

#Create a single list with all strings
        for ax,i in enumerate(x):                                                  
            y = y + x[ax]

# Created single list where strings are converted to integers
        for r in y:                                                                
            j = r.split(", ")
            j[0] = int(j[0])
            j[1] = int(j[1])
            z.append(j)
        return z

# Datasplit function - splits the data into two parts
    def datasplit_function(self, list1):                                           
        l1 = []
        l2 = []
        for i in list1[:1000]:
            l1.append(i)
        for i in list1[1000:]:
            l2.append(i)
        return l1,l2

# Mapper function - maps and combines data simultaneously
    def mapper_function(self, data):                                               
        lst = []                                                                   
        for i in data:
            lst.append((int(i[0]/100), i[1]))   
        xy = {}
        print(type(lst),"\n")
        for item in lst:
            if item[0] in xy:
                xy[item[0]] = max(xy[item[0]], item[1])
            else:
                xy[item[0]] = item[1]
        return xy

# Sort function - sorts data in ascending order
    def sort_function(self,map1,map2):                                             
        lst = {}
        for i in sorted (map1) :
            lst[i] = map1[i]
        for item in sorted(map2):
            if item in lst:
                lst[item] =  [map1[i] for i in map1 if i==item]
                lst[item].append(map2[item])
            else:
                lst[item] = map2[item]
        return lst

# Partition function - partitions year wise
    def partition_function(self,dict1):                                            
        d1 = {}
        d2 = {}
        for i in dict1:
            if i in range(2010,2016):
                d1[i] = dict1[i]
            else:
                d2[i] = dict1[i]
        return d1,d2
    
# Reducer function - finds the maximum temperature
    def reducer_function(self,bx):                                                 
        new_d = {}
        for i in bx:
            if len(bx[i])>1:
                new_d[i] = max(bx[i]) 
            else:
                new_d[i] = bx[i]
        return new_d
                
# Creating an object of the class
mapred = BDHW()                                                               

# Converted the given string data into list of integers
finaldata = mapred.converting_listofintegers(lines)                                

# Spliting the data into two parts
data1,data2 = mapred.datasplit_function(finaldata)                                 


map1 = mapred.mapper_function(data1)
map2 = mapred.mapper_function(data2)
sorted_data = mapred.sort_function(map1,map2)
reducer_function_ip1,reducer_function_ip2 = mapred.partition_function(sorted_data)
final1 = mapred.reducer_function(reducer_function_ip1)
final2 = mapred.reducer_function(reducer_function_ip2)

# Combined all the outputs
total = {**final1, **final2}                                                       

# Converted the dictionary into data frame
df = pd.DataFrame.from_dict(total,orient='index')                               
df.columns = ['Max temp']
df.reset_index(inplace=True)
df = df.rename(columns = {'index':'year'})
print(df)

# Saved the dataframe into csv file
df.to_csv('output.csv')                                                     
```
